```python
#Import pathlib and csv libraries
import csv
from pathlib import Path

#Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('/Users/TFitz/Documents/School/Starter_Code/PyRamen/Resources/menu_data.csv')
sales_filepath = Path('/Users/TFitz/Documents/School/Starter_Code/PyRamen//Resources/sales_data.csv')

#Initialize list objects to hold menu and sales data
menu = []
sales = []

#Read in the menu data
with open(menu_filepath, 'r') as menu_csv:
    csvreader = csv.reader(menu_csv, delimiter=',')
    header = next(csvreader)
    for row in csvreader:
        menu.append(row)

#Read in the sales data
with open(sales_filepath, 'r') as sales_csv:
    csvreader = csv.reader(sales_csv, delimiter=',')
    header = next(csvreader)
    for row in csvreader:
        sales.append(row)

#Initialize dict object 
report = {}

#Loop through sales data
for row in sales:
    
    #Initialize sales data variables
    quantity = int(row[3])
    sales_item = row[4] 

    #Check if sales item is in report, if not add as new key-value pair
    if sales_item not in report:
        report[sales_item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}
        
    #Loop through the menu data
    for row in menu:
        
        #Initialize menu data variables
        item = row[0]
        price = float(row[3])
        cost = float(row[4])

        #Begin tracking sales item metrics when matched with menu data
        if item == sales_item:
            profit = (price - cost)
            report[sales_item]["01-count"] += quantity
            report[sales_item]["02-revenue"] += price * quantity
            report[sales_item]["03-cogs"] += cost * quantity
            report[sales_item]["04-profit"] += profit * quantity
            break 
    else:
        print(f"{sales_item} does not equal {item}! NO MATCH!")

for key, value in report.items():
        print(f"{key} {value}\n")

#Write out report to a text file (won't appear on the command line output)
with open('report.txt', 'w') as file:
    for key, value in report.items():
        file.write(f"{key} {value}\n")
        
```
